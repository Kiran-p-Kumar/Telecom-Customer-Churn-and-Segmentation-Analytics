```python
#fix total charges 
df['Total Charges'] = pd.to_numeric(df['Total Charges'], errors = 'coerce')
```

```python
df[['Total Charges']]
```

```python
#Tenure Groups 
def tenure_groups (tenure):
    if tenure <= 12:
        return '0-12 Months'
    elif tenure <= 24:
        return '13-24 Months'
    elif tenure <= 48:
        return '25-48 Months'
    else: 
        return '48+ Months'
df['Tenure Groups'] = df['Tenure in Months'].apply(tenure_groups)
        
```

```python
df[['Tenure Groups']]
```

```python
#initial risk 

def risk_level (row):
    if row['Contract'] == 'Month-to-Month' and row['Monthly Charge'] > 70:
        return 'High Risk'
    elif row['Contract'] == 'One year':
        return 'Medium Risk'
    else:
        return 'Low Risk'
df['Risk Level'] = df.apply(risk_level,axis = 1)
```

```python
df[['Risk Level']]
```

```python
#Churned flag 
df['Churned_Flag'] = df['Customer Status'].apply(lambda x: 1 if x == 'Churned' else 0)
```
